### Types of Machine learning
- Supervised
   
   We provide the algorithm with the inputs and the desired outputs
   Based on those details it learns to produce outputs close to the ones we desire.
   
   Classification Models - Cows /bulls
   
   Regression Models- numerical types,predicting stocks


- Unsupervised

  We feed inputs but there are nopt target output.
  For example labelling pictures in a data set, we provide the algorith with pictures labeled ie. cow,bull and the algorithm trains its self until it can identify a bull or cow on its own.
  Usually used when we want to split the data into certain number of categories which we do know prior to implementing it.
  Clustering
  
  
- Reinforcement

  We train a model to act in an environment  based on the rewards it will receive.
  Much like traing a pet to do certian things and reward it later whenever it performs well.
   ie. Training an algorithm to play super Mario.
